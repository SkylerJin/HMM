```python
#### run this cell
import numpy as np
import pandas as pd
from hmmlearn import hmm
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.formula.api as smf
import statsmodels.api as sm
import seaborn as sns
from scipy import stats
sns.set(style='ticks', color_codes=True)
```

```python
df = pd.read_csv('IBM_exp.csv', sep=',')
```

```python
df.head()
```

```python
rows, columns = df.shape
X = df.iloc[:,1:columns]
gap = X.Close - X.Open
dif = X.High - X.Low
vol = X.Volume
X = np.column_stack([dif, gap, vol])
```

```python
res = []
model = hmm.GaussianHMM(6, covariance_type = "diag", n_iter = 100)
for i in range(100):
    X_train = X[i:i+100]
    model.fit(X)
    hidden_state_test = model.predict(X_train)
    state = hidden_state_test[99]
    state_next = np.argmax(model.transmat_[state])
    mean = model.means_[state_next]
    cov = model.covars_[state_next]
    change = np.random.multivariate_normal(mean, cov)
    pred_train = change[1]
    res.append(pred_train)
```

```python
y_true = df.Close[100:200]
y_true.head()
```

```python
plt.plot(range(100),y_true, "-", c="red",label='True')
plt.plot(range(100),res+df.Open[100:200], "-", c="b")
plt.xlabel('Index')
plt.ylabel('Close Price')
plt.savefig('price.png', dpi=1000)
```

```python
plt.plot(range(100), gap[100:200], "-", c="red",label='True')
plt.plot(range(100), res, "-", c="b",label = 'Predict')
plt.xlabel('Index')
plt.ylabel('Diffence of Close Price and Open Price')
plt.savefig('diff.png', dpi=1000)
```
